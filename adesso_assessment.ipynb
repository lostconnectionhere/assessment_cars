## Introduction
In this jupyter notebook I will explore the data that was provided by Adesso. The main goal of this assessment is to develop a predictive model capable of determining in advance whether a car will sustain damage.

### The dataset
The dataset contains different characteristics of cars. 

1. PSEUDO_VIN: Encrypted chassis number for identification of individual instances
2. TONDUR: A variant of an operating hours counter
3. EngSpDist_K{1,14}: Distribution of dwell time in 14 disjoint regions of the engine speed range; engine idling takes place in the range of class K1
4. AmbTempDist_K{1,10}: Distribution of residence time in 10 disjoint ranges of outdoor temperature; temperature increases from K1 to K10
5. KMSTAND: Mileage of the vehicle at the considered readout
6. AUSLESEDATUM: The date of the considered readout
7. REPARATURDATUM: Date of the executed repaid (if present, the vehicle has the label "DAMAGE").
8. TACHOMETER_IN_KM: Mileage at the time of the performed repair (if present, the vehicle has the label "DAMAGE").
9. LABEL: Determined class affiliation based on repair history; target variable for binary classification.
